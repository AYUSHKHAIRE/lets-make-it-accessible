```python
import requests
from bs4 import BeautifulSoup
```

```python
query = 'ml+engineer'
url = f"https://www.google.com/search?q={query}"
```

```python
# Set headers to mimic a browser visit
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
}
```

```python
# Send a GET request to fetch the search results
response = requests.get(url, headers=headers)
# Parse the HTML content using BeautifulSoup
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
print(soup.prettify())
```

```python
results = soup.find_all('div', {'class': 'MjjYud'})
for r in results:
    # results snipet
    snipet = r.find(
        'div', {'class': 'kb0PBd cvP2Ce A9Y9g'})
    snipet2 = r.find(
        'span', {'class': 'hgKElc'})
    if snipet:
        print("Snippet:", snipet.get_text())
    if snipet2:
        print("Snippet:", snipet2.get_text())
    # links
    a_s = r.find_all('a')
    for a in a_s:
        if 'search' in a['href']:
            continue
        else:
            print(a['href'])
    # provider
    snipet3 = r.find(
        'span', {'class': 'VuuXrf'})
    if snipet3:
        print(snipet3.get_text())
    print('\n\n\n')
```
